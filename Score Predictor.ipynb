# We will now convert the textual data to numeric data so that those columns can be used for prediciton

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['bat_team']=le.fit_transform(df['bat_team'])
df['venue']=le.fit_transform(df['venue'])
df['bowl_team']=le.fit_transform(df['bowl_team'])
df['batsman']=le.fit_transform(df['batsman'])
df['bowler']=le.fit_transform(df['bowler'])
```

```python
df.head(15)
```

# Dividing Dependent and Independent Variables

```python
x = df.iloc[:,[1,2,3,4,5,6,7,8,11,12]].values
y = df.iloc[:, 13].values
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

# Scaling the Dependent and Independent Variables

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Linar Regression

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(x_train,y_train)
```

# Testing the Accuracy

```python
# Testing the dataset on trained model
from sklearn.metrics import r2_score,accuracy_score
y_pred = lin.predict(x_test)
score = lin.score(x_test,y_test)
print("R square value:" , score)
```

```python
def custom_accuracy(y_test,y_pred,thresold):
    right = 0

    l = len(y_pred)
    for i in range(0,l):
        if(abs(y_pred[i]-y_test[i]) <= thresold):
            right += 1
    return ((right/l)*100)
print("Custom accuracy:" , custom_accuracy(y_test,y_pred,20))
```

# Test Case 

```python
new_prediction = lin.predict(sc.transform(np.array([[100,12,2,391,283,50,4,5.2,11,12]])))
print("Prediction score:" , new_prediction)
```
